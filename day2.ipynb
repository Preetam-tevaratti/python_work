Create arrays

```python
import numpy as np
a = np.array([1, 2, 3, 4, 5])
b = np.arange(1, 6)
c = np.zeros((3))
d = np.ones((3, 2))

print("Array a:", a)
print("Array b:", b)
print("2x3 Zeros:", c)
print("3x2 Ones:", d)
```


Indexing & Slicing

```python
print("First element of a:", a[0])
print("Last two elements of a:", a[-2:])
```


Basic arithmetic operations

```python
print("a + 2:", a + 2)
print("a * 3:", a * 3)
```

Basic statics

```python
print("Square of a:", np.square(a))
print("Mean of a:", np.mean(a))
print("Standard deviation:", np.std(a))
```

Reshaping arrays

```python
matrix = np.arange(1, 10).reshape((3, 3))
print("Reshaped 3x3 matrix:")
matrix
```

Transpose

```python
print("transpose:")
matrix.T
```

Bolean indexing

```python
print("Element > 5:", matrix[matrix > 5])
```
